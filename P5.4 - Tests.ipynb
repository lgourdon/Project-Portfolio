## Shapiro des variables

Shapiro-Wilk est un test de normalité.<br>
Il est indiqué pour le traitement d'échantillons réduits<br>
Il retourne les valeurs W = test statistique et p-value = seuil hypothèse loi normale

**Interprétation**<br>
* Hypothèse nulle H0 : la variable suit une loi normale - seuil α = 5% 
<br>
* Si p-value < α : l'hypothèse nulle est rejetée 
* Si la p-value > α : on ne doit pas rejeter l'hypothèse nulle

### Cluster 1

```python
# rappel du cluster 1
cl1= pd.read_csv('data/cl1.csv', index_col = 0)
cl1.drop(['clust'], axis=1, inplace = True)
cl1.head(1)
```

#### Test de Shapiro

```python
# Test Shapiro dans le cluster 1
for column in cl1:
    a = stats.shapiro(cl1[column])
    print(column, " : ", a)
```

## Tests de comparaison

```python
# Rappel du test de Shapiro sur Variable 'edb_2019' pour les clusters 1 et 5
shapcl1=stats.shapiro(cl1['edb_2019'])
shapcl5=stats.shapiro(cl5['edb_2019'])

print("p-value de 'edb_2019' (test de Shapiro) pour le cluster 1 =", shapcl1[1])
print("p-value de 'edb_2019' (test de Shapiro) pour le cluster 5 =", shapcl5[1])
```

### Test de Bartlett

* Comparation des variances avec le Test de Bartlett
* Hypothèse nulle (H0) = Egalité des variances de la variable 'edb_2019' sur les groupes 1 et 5

```python
b = stats.bartlett(cl1['edb_2019'], cl5['edb_2019'])
print("p-value (test de Bartlett) pour la variable 'edb_2019' =", b[1])
```

**Hypothèse nulle rejetée**<br>
* Les conditions de validité du ttest ne sot pas remplies
* Le Student test sur l'égalité des moyennes n'est pas nécéssaire<br>
* Toutefois, nous terminons la démonstration

### t-test

```python
t = stats.ttest_ind(cl1['edb_2019'], cl5['edb_2019'], equal_var=True)
print("p-value t-test de la variable 'edb_2019' sur les clusters 1 et 5 =", t[1])
```

**Hypothèse nulle rejetée**
* Différence significative de la moyenne, conditions non valides

### Test de Wilcoxon

Le test des rangs signés de Wilcoxon est une alternative **non-paramétrique** au test de Student pour des échantillons appariés.<br> 
Il s'intéresse à un paramètre de position : la médiane, le but étant de tester s'il existe un changement sur la médiane.

Le test de Wilcoxon sur 2 samples est aussi appelé test Mann-Whitney

```python
from scipy.stats import mannwhitneyu

# Comparation des samples
stat, p = mannwhitneyu(cl1['edb_2019'], cl5['edb_2019'])
print('Statistics=%.3f, p=%.3f' % (stat, p))

# interprétation
alpha = 0.05
if p > alpha:
	print("Même distribution (l'hypothèse H0 ne peut pas être rejetée)")
else:
	print("Distribution différente (rejet de H0)")
```

**Hypothèse nulle rejetée**

Les distributions sont bien différentes
